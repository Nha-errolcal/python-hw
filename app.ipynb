```python

def cube():
    a = float(input("please input number A"))
    s = round(6 * a**2 ,2)
    print(f"Surface area = {s}")

    v = round(a**3,2)
    print(f"Volume = {v}")

    d = round(a*(3**0.5),2)
    print(f"Diagonal = {d}")

def rectangularBox():
    a = float(input("Please input number A: "))
    b = float(input("Please input number B: "))
    c = float(input("please input number C: "))
    
    s = round( 2*(a*b + (a*c) + (b*c)),2)
    print(f"Surface = {s}")

    v = round(a * b *c ,2)
    print(f"Volume = {v}")
    
    d = round(a**2 + b**2 + c**2 * 0.5,2)
    print(f"Diagonal = {d}")

def  cylinder():
    r = float(input("please input number R: "))
    h = float(input("please input number H: "))
    
    sL = round(2 * 3.14* r *h,2)
    print(f"Lateral surface = {sL}")
    
    sB = round(3.14 * r**2,2)
    print(f"Base = {sB}")
    
    tS = round(sL + 2*sB)
    print(f"Total surface = {tS}")

    volume = round(3.14 * r**2*h,2)
    print(f"Volume = {volume}")
    
def menu():
    print("\n==== MENU ====")
    print("1. Cube")
    print("2. Rectangular Box")
    print("3. Cylinder")
    print("4. Exit")
    print("\n=============")

def main():
    while True:
        menu()
        ch = input("Enter your choice: ")

        if ch == '1':
            cube()
        elif ch == '2':
            rectangularBox()
        elif ch == '3':
            cylinder()
        elif ch == '4':
            print("Goodbye ðŸ‘‹")
            break
        else:
            print("Invalid choice, try again!")

main()
```
