```python
# ============== task 3 ==============

def average(num1, num2, num3):
    return (num1 + num2 + num3) / 3


a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

result = average(a, b, c)
print("Average =", result)
```
