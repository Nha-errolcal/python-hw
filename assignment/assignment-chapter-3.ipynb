```python
# task 1

import numpy as np

print("========== TASK 1 ==========")

arr = np.random.randint(1, 51, 15)

print("Array:", arr)
print("Sum:", arr.sum())
print("Mean:", arr.mean())
print("Min:", arr.min())
print("Max:", arr.max())
print("Standard Deviation:", arr.std())
```

```python
print("\n========== TASK 2 ==========")

matrix = np.random.randint(1, 10, (4, 4))

print("Matrix:\n", matrix)

det = np.linalg.det(matrix)
print("Determinant:", det)

print("Transpose:\n", matrix.T)
```
