```python
import matplotlib.pyplot as plt
import pandas as pd

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
monthly_sales = [1200, 1500, 1700, 1600, 1900, 2100]

products = ["Laptop", "Phone", "Tablet", "Headset"]
product_sales = [3000, 5200, 2100, 1400]

plt.figure(figsize=(8, 4))
plt.plot(months, monthly_sales, marker="o", label="Monthly Sales")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
plt.bar(products, product_sales, label="Product Sales")
plt.title("Product Sales")
plt.xlabel("Product")
plt.ylabel("Sales (USD)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "Student": ["Sok", "Panha", "Vanna", "Kanha", "Rith"],
    "Score": [78, 92, 85, 90, 88]
})

max_score = df["Score"].max()
top_student = df[df["Score"] == max_score]["Student"].values[0]

plt.figure(figsize=(8, 4))
bars = plt.bar(df["Student"], df["Score"])

for i, (student, score) in enumerate(zip(df["Student"], df["Score"])):
    if score == max_score:
        bars[i].set_hatch("//") 
        plt.text(i, score + 1, f"Top: {student} ({score})", ha="center")

plt.title("Student Scores")
plt.xlabel("Student")
plt.ylabel("Score")
plt.ylim(0, 100)
plt.tight_layout()

plt.gcf()
plt.show()

print("Highest score:", max_score, "by", top_student)
print(" Saved: assignment2_student_scores.png")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    "Student": [f"Student_{i}" for i in range(1, 21)],
    "Math": np.random.randint(40, 100, 20),
    "Science": np.random.randint(40, 100, 20),
    "English": np.random.randint(40, 100, 20)
}
df = pd.DataFrame(data)

df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)

top10 = df.sort_values("Average", ascending=False).head(10)

plt.figure(figsize=(9, 4))
plt.bar(top10["Student"], top10["Average"])
plt.title("Top 10 Students by Average Score")
plt.xlabel("Student")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
plt.hist(df["Average"], bins=10)
plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.tight_layout()
plt.show()

subject_means = df[["Math", "Science", "English"]].mean()

plt.figure(figsize=(7, 4))
plt.plot(subject_means.index, subject_means.values, marker="o")
plt.title("Average Score per Subject")
plt.xlabel("Subject")
plt.ylabel("Mean Score")
plt.grid(True)
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].plot(subject_means.index, subject_means.values, marker="o")
ax[0].set_title("Mean Score per Subject")
ax[0].set_xlabel("Subject")
ax[0].set_ylabel("Mean Score")
ax[0].grid(True)

top5 = df.sort_values("Average", ascending=False).head(5)
ax[1].bar(top5["Student"], top5["Average"])
ax[1].set_title("Top 5 Students (Average)")
ax[1].set_xlabel("Student")
ax[1].set_ylabel("Average Score")
ax[1].tick_params(axis="x", rotation=45)

plt.tight_layout()

plt.savefig("assignment34exam_project.png", dpi=300)
plt.show()
```
