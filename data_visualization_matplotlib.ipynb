```python
import matplotlib.pyplot as plt

product = ["A","B","C"]
salary = [90,100,80]
plt.pie(salary,labels=product, autopct='%1.1f%%')
plt.show()
```

```python
import numpy as np

data = np.random.randn(10000)
plt.hist(data, bins=20, color="pink", edgecolor="black")
plt.title("Basic histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python

days = [1, 2, 3, 4, 1]
timep = [2, 4,-16, 8, 10]
plt.figure(figsize=(16,7))

plt.plot(days, timep, color='blue')
plt.grid(True)
plt.title("Sports watch data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burmage")
plt.show()
```

```python
import matplotlib.pyplot as plt1
subject = ["Computer","English","Physics","biology","math","history"]
myMark = [100,90,80,93,100,80]
friendMark = [80,70,100,99,64,90]
plt1.figure(facecolor="pink",figsize=(15,7))
plt1.plot(subject,myMark,marker='o',label="My work")
plt1.plot(subject,friendMark,marker='o',label="My Friend's")
plt1.legend()
plt1.title("mark comparison")
plt1.xlabel("Subject")
plt1.ylabel("Mark")
plt1.grid(True)
plt1.show()
```
