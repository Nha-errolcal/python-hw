```python
import pandas as pd
import numpy as np

#1
data = {
    "A": np.random.randint(1, 100, 10),
    "B": np.random.randint(1, 100, 10),
    "C": np.random.randint(1, 100, 10),
    "D": np.random.randint(1, 100, 10),
    "E": np.random.randint(1, 100, 10)
}

df = pd.DataFrame(data)

print("First 5 rows:")
print(df.head())
```

```python
#2
filtered_df = df[df["A"] > 50]

print("\nRows where A > 50:")
print(filtered_df)
```

```python
#3
df["Sum_AB"] = df["A"] + df["B"]

print("\nDataFrame with new column:")
print(df.head())
```

```python
#4
df.loc[0] = df.loc[1] # duplicate
df.loc[2, "A"] = np.nan # missing value

df = df.drop_duplicates()

df = df.fillna(df.mean(numeric_only=True))

print("\nAfter cleaning:")
print(df.head())
```

```python
#5
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Score": [85, 90, 95]
})

merged_df = pd.merge(df1, df2, on="ID")

print("\nMerged DataFrame:")
print(merged_df)
```

```python
#6
import pandas as pd
import numpy as np
dates = pd.date_range(start="2024-01-01", periods=90, freq="D")
ts = pd.DataFrame({
    "Date": dates,
    "Value": np.random.randint(100, 200, 90)
})

ts.set_index("Date", inplace=True)

monthly = ts.resample("ME").mean()

print("\nMonthly Resampled Data:")
print(monthly)
```
