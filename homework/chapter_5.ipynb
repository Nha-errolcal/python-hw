```python
#1
import matplotlib.pyplot as plt
import numpy as np

days = []
temperature = [30, 32, 31, 29, 28, 33, 34]

for day in range(1, 8):
    days.append(day)

plt.plot(days, temperature, marker='o')
plt.title("Temperature Over 7 Days")
plt.xlabel("Day")

plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.show()
```

```python
#2
subjects = ["Math","Science", "English"]
marks = [80,75,90]

plt.bar(subjects,marks)
plt.title("Student Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
#3

random_data = np.random.randn(1000)

plt.hist(random_data, bins=15)
plt.title("Histogram 15 Bins")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#4
brands = ["Apple", "Samsung", "Xiaomi", "Oppo", "Vivo"]
market_share = [30, 25, 20, 15, 10]

plt.pie(market_share, labels=brands, autopct="%1.1f%%")
plt.title("Mobile Market Share")
plt.show()
```

```python
#5
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(days, temperature, marker='o')
ax[0].set_title("Temperature")
ax[0].set_xlabel("Day")
ax[0].set_ylabel("Â°C")

ax[1].bar(subjects, marks)
ax[1].set_title("Marks")
ax[1].set_xlabel("Subjects")
ax[1].set_ylabel("Score")

plt.tight_layout()
plt.show()
```
