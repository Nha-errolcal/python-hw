```python
import numpy as np

def task1():
    arr = np.random.randint(1,51, size=15)
    print("Main Array Task 1:", arr)
    
    sum = np.sum(arr)
    mean = np.mean(arr)
    min = np.min(arr)
    max = np.max(arr)
    std = np.std(arr)
    
    print(f"Sum: {sum}")
    print(f"Mean: {mean:.2f}")
    print(f"Min: {min}")
    print(f"Max: {max}")
    print(f"Standard deviation: {std:.2f}")

def task2():
    arr = np.array([[1, 2, 3, 7],
                    [0, 1, 4, 9],
                    [5, 6, 16, 12],
                    [6, 2, 8, 1]
                    ])
    print("Main Array Task 2:\n", arr)
    
    determ = np.linalg.det(arr)
    print(f"determ: {determ:.2f}")
    
    print(f"\ntranspose:\n{arr.T}")

    invArr = np.linalg.inv(arr)
    print(f"\ninverse:\n{invArr}")
    

def task3():
    arr = np.random.randint(1, 101, size=10)
    
    shuffled_array = arr.copy()
    np.random.shuffle(shuffled_array)
    print("shuffle:", shuffled_array)
    
    print(f"original array:", arr)
    
    
def main():
    print("----- Task 1 -----")
    task1()
    print("\n----- Task 2 -----")
    task2()
    print("\n----- Task 3 -----")
    task3()

main()
```
