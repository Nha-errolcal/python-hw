```python
import pandas as pd


# List
data = {
    "Name":["PL","LP","GJ","YY","RR"],
    "Age":[22,55,23,24,19],
    "Salary":[500,777,555,999,222]
}

df = pd.DataFrame(data)
ages = df["Age"]
subSet = df[["Name","Salary"]]
filterAge = df[df["Age"] == 22]
filterSalary = df[df["Salary"] >= 230]

print(f"Age :\n{ages}")
print(f"SubSet :\n{subSet}")
print(f"filterAge :\n{filterAge}")
print(f"filterSalary :\n{filterSalary}")
```

```python
#data cleaing
import pandas as pd

data1 = {
    "name":["KOK1","KOK","POPO"],
    "salary":[1000,2000,800],
    "age":[20,27,18]
}

df1 = pd.DataFrame(data1)
df1["Major"] = ["CS","Network","DS"] #adding
df1["year"] = ["y1","y2","y3"] #adding
print(f"After adding filed {df1}")


dup = df1.replace("NET","Cs")
print(f"{dup}")

#convert to data types
type=df1["age"].astype(float)
print(type)
```

```python
# sort
print(f"Before sorting\n{df1}")

sorts = df1.sort_values("salary", ascending=False)

print(f"After sorting\n{sorts}")
```

```python
# aggregation

# Sum salary by Major
sum_salary = df1.groupby("Major")["salary"].sum()
print(f"Sum value = \n{sum_salary}\n")

# Count Major by year
count_major = df1.groupby("year")["Major"].count()
print(f"Count value = \n{count_major}")
```
